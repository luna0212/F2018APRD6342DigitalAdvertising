# Travel Pony Ad Cost Regression Analysis

```python
import pandas as pd
import statsmodels.api as sm
```

```python
#import data
df = pd.read_csv("/Users/pudin/Desktop/digital/Travel Pony Facebook.csv")
df[df['Amount Spent (USD)'] > 0]
```

```python
#multiple regression

df['Campaign Name'] = df['Campaign Name'].astype('category') #convert
df['Campaign Name'] = df['Campaign Name'].cat.codes #convert

X = df[['Campaign Name','Reach','Frequency','Impressions',
'Clicks','Unique Clicks','Page Likes','Page Engagement',
'Post Engagement','Post Likes','Post Comments',
'Post Shares','Photo Views','Website Clicks']] 
y = df['Amount Spent (USD)'] 
X = sm.add_constant(X) #add an intercept 

model = sm.OLS(y.astype(float), X.astype(float)).fit()
predictions = model.predict(X)

print(model.summary())

model.params.sort_values(ascending=False)
```
