Construct the data set that generates the scatter plot from the website below. Data sourse is FRED.

https://economicsone.com/2011/01/14/higher-investment-best-way-to-reduce-unemployment-recent-experience-shows/



```r
#data cleanning

p1 <- read.csv("~/Desktop/MSBC5030 QM/Project2/GCE.csv")
p2 <- read.csv("~/Desktop/MSBC5030 QM/Project2/GDP.csv")
p3 <- read.csv("~/Desktop/MSBC5030 QM/Project2/UNRATE.csv")

p1$DATE <- as.Date(as.character(p1$DATE), format = '%Y-%m-%d')
p2$DATE <- as.Date(as.character(p2$DATE), format = '%Y-%m-%d')
p3$DATE <- as.Date(as.character(p3$DATE), format = '%Y-%m-%d')

df0 <- merge(x=p1,y=p2,by="DATE")
df <- merge(x=df0,y=p3,by="DATE")
```

```r
#Create plot

df$GP_as_percent_GDP <- df$GCE / df$GDP *100

df2 <- df[substr(df$DATE,1,4) >= "1990" & substr(df$DATE,1,4) < "2010",]
a <- plot(df2$GP_as_percent_GDP,df2$UNRATE,xlab="Government purchases as a percent of GDP",ylab="Unemployment rate",ylim=c(3,11),xlim=c(17.0,22.0),pch=16,col='blue')


df3 <- df[substr(df$DATE,1,4) >= "1948" & substr(df$DATE,1,4) <= "1990",]
b <- plot(df3$GP_as_percent_GDP,df3$UNRATE,xlab="Government purchases as a percent of GDP",ylab="Unemployment rate",pch=16,col='red')

df4<-df[substr(df$DATE,1,4) >= "2010" & substr(df$DATE,1,4) <= "2018",]
plot(df2$GP_as_percent_GDP,df2$UNRATE,xlab="Government purchases as a percent of GDP",ylab="Unemployment rate",ylim=c(3,11),xlim=c(17.0,22.0),pch=15,col=rgb(1,0,0,.6))
points(df3$GP_as_percent_GDP,df3$UNRATE,pch=16,col=rgb(0,1,0,.5))
points(df4$GP_as_percent_GDP,df4$UNRATE,pch=17, col=rgb(0,0,1,.5))
legend("topleft", 
       legend = c("1948-1990","1990-2010", "2010-2018"), 
       col = c(rgb(0,1,0,.5), 
               rgb(1,0,0,.6), rgb(0,0,1,.5)),
       pch = c(16,15,17), 
       bty = "n", 
       pt.cex = 1.4, 
       cex = 0.8, 
       text.col = "black", 
       horiz = F)
```
